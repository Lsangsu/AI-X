## 3-8. 피벗 테이블(pivot table) 기능과 그룹 통계(groupby)

```python
df = pd.read_csv('kpop_idols.csv')
df = df.iloc[:200, 1:11]
df
```

```python
# 키와 몸무게가 0이 아닌 값만 사용

df = df.loc[(df['Height']!=0)&(df['Weight']!=0)]
df
```

```python
# 그룹 이름이 없는 행 삭제

df = df.dropna(subset = ['Group'], axis = 0)
df
```

```python
# 인덱스 번호 재정립

df = df.reset_index(drop = True)
df
```

### 3-8-1. 피벗 테이블(pivot_table)

```python
# index = '행 인덱스', columns = '열 인덱스', values = '조회하고 싶은 값', aggfunc = '집계방식(기본값은 평균)'

pd.pivot_table(df, index='Group', columns='Country', values='Height', aggfunc='mean')
```

```python
pd.pivot_table(df, index='Group', columns='Country', values='Height', aggfunc='count')
```

### 3-8-2. 그룹 통계(groupby)

```python
len(df)
```

단일 그룹 지정

```python
df.groupby('Group').mean(numeric_only = True)   
# Group별로 묶어서 numeric_only(숫자 데이터만)
```

```python
df.groupby('Group').sum(numeric_only = True)   
```

```python
df.groupby('Group')['Height'].mean()   # 특정 컬럼 지정
```

여러 개 그룹 지정

```python
df.groupby(['Group','Country']).mean(numeric_only = True)
```

```python
# 인덱스 분할 해제 : reset_index()

df.groupby(['Group','Country']).mean(numeric_only = True).reset_index()
```

그룹 데이터 확인 : get_group()

```python
df.groupby('Group')
```

```python
# 그룹으로 묶인 목록 확인

df.groupby('Group').groups
```

```python
df.groupby('Group').get_group('AoA')
```

그룹 집계 방식 구분하기 : agg()

```python
df.groupby('Group').agg({'Height':'mean', 'Weight':'sum'}).reset_index()
```

```python
df.groupby('Group').agg({'Height': 'mean', 'Weight' : 'sum'}).round(2).reset_index()
```
