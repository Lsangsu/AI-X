## 3-9. 데이터 합치기(Join, Merge)

```python
import pandas as pd
```

```python
df = pd.read_csv('kpop_idols.csv')
df
```

```python
# 2개의 데이터프레임 생성

df1 = df.iloc[:5, 1:11]
df2 = df.loc[3:10, ['Korean Name', 'Instagram']]
```

```python
df1
```

```python
df2
```

### 3-9-1. Join : 인덱스 기준 합치기

```python
# join을 사용하기 위해서, 키로 지정할 열을 인덱스화 : set_index('키로 지정할 열')

df2.set_index('Korean Name')
```

1. Left join

```python
df1.join(df2.set_index('Korean Name'), on='Korean Name', how='left')
```

2. Right join

```python
df1.join(df2.set_index('Korean Name'), on = 'Korean Name', how = 'right')
```

3. Inner join

```python
df1.join(df2.set_index('Korean Name'), on='Korean Name', how = 'inner')
```

4. Outer join

```python
df1.join(df2.set_index('Korean Name'), on = 'Korean Name', how = 'outer')
```

```python
df1.join(df2.set_index('Korean Name'), on = 'Korean Name', how = 'outer').reset_index()
```

### 3-9-2. Merge : 겹치는 값 기준으로 합치기

1. left merge

```python
pd.merge(df1, df2, on = 'Korean Name', how = 'left')
```

```python
df1.merge(df2, on = 'Korean Name', how = 'left')
```

2. right merge

```python
pd.merge(df1, df2, on = 'Korean Name', how = 'right')
```

3. inner merge

```python
df1.merge(df2, on = 'Korean Name', how = 'inner')
```

4. outer merge

```python
df1.merge(df2, on = 'Korean Name', how = 'outer')
```

### 3-9-3. Concat : 단순히 데이터 붙이기

1. 행 방향으로 붙이기 : axis=o

```python
pd.concat([df1, df2], axis = 0)
```

2. 열 방향으로 붙이기 : axis=1

```python
pd.concat([df1, df2], axis =1)
```

3. 겹치는 값만 붙이기(join 활용)

```python
pd.concat([df1, df2], axis=1, join = 'inner')
```
