## 3-6-4. 행, 열 추가 및 삭제

```python
df
```

```python
# 딕셔너리를 활용한 행 추가

new_row = {'Stage Name': 'FastCampus', 
           'Full Name': 'Fast Campus', 
           'Korean Name' : '패스트캠퍼스', 
           'K. Stage Name' : '패캠', 
           'Date of Birth' : '2017-01-01',
           'Group' : 'FC',
           'Country' : 'Soutch Korea',
           'Second Country' : None,
           'Height' : 180,
           'Weight' : 80}
```

```python
# new_row를 데이터프레임으로 변환
new_row_df = pd.DataFrame([new_row])

# concat 사용
df = pd.concat([df, new_row_df], ignore_index=True)
```

```python
df
```

```python
# 시리즈를 활용한 추가

new_row2 = pd.DataFrame([df.iloc[0]])
df = pd.concat([df, new_row2], ignore_index=True)
df
```

```python
# 열 추가

df['Blood Type'] = 'A'
df
```

```python
df['Blood Type'] = ['A']*5 + ['B']*4 + ['O']*2 + ['AB']*1
df
```

```python
# 행 삭제
df = df.drop(index=[10, 11])
df
```

```python
# 열 삭제

df = df.drop(columns='Blood Type')
df
```

```python
df = df.drop('Second Country', axis = 1)
```
