## 3-12. apply, lambda, map 함수

```python
# 라이브러리 임포트 / 파일 로딩
import pandas as pd
import numpy as np

df = pd.read_csv('kpop_idols.csv')
df
```

```python
df.info()
```

```python
df.loc[df['Gender'] == 'M']
```

```python
# Gender가 M인 상위 3개 데이터 + 원본 데이터프레임 상위 9개 결합

df = pd.concat([df.loc[df['Gender']=='M'][:3], df.head(9)]).reset_index(drop=True)
df
```

```python
df = df[['Weight', 'Height', 'Gender']]
df
```

### 3-12-1. apply 함수 : 로직이 정의된 함수를 만들고, 그 함수를 적용할 때 apply() 함수에 대입

```python
# Gender 기준 키, 몸무게 평균(결측치 제외)

df.replace(0, np.nan).dropna().groupby('Gender').mean()
```

```python
def replace_zero(x):
    if x == 0:
        x = df['Height'][3:11].mean()
    return x
```

```python
df['Height'] = df['Height'].apply(replace_zero)
df
```

### 3-12-2. lambda 함수

```python
df['Weight'] = df['Weight'].apply(lambda x: df['Weight'][3:11].mean() if x == 0 else x)
df
```

```python
# 'Weight' / 'Height' 계산괸 컬럼 만들기

df['Wight/Height'] = df.apply(lambda x: x['Weight'] / x['Height'], axis=1)
df
```

### 3-12-3. map 함수

```python
df
```

```python
map_info = {'M' : '남자',
            'F' : '여자'}

df['Gender'] = df['Gender'].map(map_info)
df
```
