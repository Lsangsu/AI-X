## 3-11.결측값, 중복 처리

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('kpop_idols.csv')
df
```

```python
df = df.iloc[15:30]
df = df[['K. Stage Name', 'Date of Birth', 'Group', 'Height', 'Weight', 'Birthplace']]
df
```

### 3-11-1. 결측값 처리

### 1. 데이터 탐색

```python
df
```

```python
# 0값을 NaN 형으로 변환
df = df.replace(0, np.nan)
df
```

```python
df.info()
```

### 2. 결측 데이터 확인 : isnull() / notnull()

```python
df.isnull()
```

```python
df.notnull()
```

```python
#  결측값 개수 확인
df.isnull().sum()
```

### 3. 결측 데이터 제거 : dropna()

```python
# 결측값이 있는 행 모두 삭제

df.dropna()
```

```python
# 특정 컬렘 지정 -> 해당 컬럼에서만 삭제
df.dropna(subset=['Group'])
```

```python
df.dropna(subset=['Group', 'Height'])
```

### 4-1. 결측 데이터 치환 : fillna()

```python
df = df.dropna(subset=['Group'])
df
```

```python
# 지정 값으로 대체 : fillna('대체값)

df['Height'] = df['Height'].fillna(160)

df
```

```python
# 평균 값으로 대체 : fillna('평균값')

weight_mean = df['Weight'].mean()
df['Weight'] = df['Weight'].fillna(weight_mean)
df
```

```python
# 최빈값으로 대체 : fillna('최빈값')
# 최빈값 확인 : value_counts()

df['Birthplace'].value_counts()
```

```python
df['Birthplace'].value_counts().index
```

```python
df['Birthplace'].value_counts().index[0]
```

```python
df['Birthplace'] = df['Birthplace'].fillna(df['Birthplace'].value_counts().index[0])
df
```

```python
df = df.reset_index(drop=True)   # drop=True : 이전의 인덱스값을 버림
df
```

### 3-11-2. 중복값 처리

```python
# 중복 데이터 임의로 생성

df = pd.concat([df, pd.DataFrame([df.iloc[-1]])], ignore_index=True)
df
```

1. 중복 데이터 탐색 : duplicated()

```python
# 중복된 행 찾기
df[df.duplicated()]
```

```python
# 중복된 열 찾기
df['Group'].duplicated()
```

2. 중복 데이터 제거하기 : drop_duplicates()

```python
# 행 기준으로 제거
df.drop_duplicates()
```

```python
# 열 기준으로 제거
df.drop_duplicates(subset=['Group'])
```

```python
# 특정 열을 기준으로 제거 (마지막 값 남기기)
df.drop_duplicates(subset= ['Group'], keep='last')
```
