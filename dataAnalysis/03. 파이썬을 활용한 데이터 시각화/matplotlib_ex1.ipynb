```python
x = np.linspace(-np.pi, np.pi, 200)
y1, y2 = np.cos(x), np.sin(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

범례 설정

```python
x = np.linspace(-np.pi, np.pi, 200)
y1, y2 = np.cos(x), np.sin(x)

plt.plot(x, y1, label='코사인 그래프')
plt.plot(x, y2, label='사인 그래프')
plt.legend(loc='upper left')
plt.show()
```

## 4-2-4. 다중 그래프

```python
# 그래프 2개 플롯팅
x1 = np.array([10, 20, 30, 40, 50, 60, 70, 80])
x2 = np.array([10, 20, 30, 40, 50, 60, 70, 80])
y1 = np.array([1, 3, 5, 7, 9, 11, 13, 15])
y2 = np.flip(y1)

plt.figure(figsize=(10, 10)) # figure 생성

ax1 = plt.subplot(2, 1, 1)  # ax1 은 2x1 에서 첫번째 그래프
plt.plot(x1, y1, 'yo-')
plt.title('첫번째 그래프')

ax2 = plt.subplot(2, 1, 2)  # ax2 은 2x1 에서 두번째 그래프
plt.plot(x2, y2, 'r.-')
plt.title('두번째 그래프')

plt.show()
```

```python
# 그래프 4개 플롯팅
x1 = np.array([10, 20, 30, 40, 50, 60, 70, 80])
x2 = np.array([10, 20, 30, 40, 50, 60, 70, 80])
y1 = np.array([1, 3, 5, 7, 9, 11, 13, 15])
y2 = np.flip(y1)

x3 = np.linspace(-np.pi, np.pi, 200)
x4 = np.linspace(-np.pi, np.pi, 200)
y3 = np.cos(x3)
y4 = np.sin(x3)

plt.figure(figsize=(10, 10)) # figure 생성

ax1 = plt.subplot(2, 2, 1)  # ax1 은 2x2 에서 첫번째 그래프
plt.plot(x1, y1, 'yo-')
plt.title('첫번째 그래프')

ax2 = plt.subplot(2, 2, 2)  # ax2 은 2x2 에서 두번째 그래프
plt.plot(x2, y2, 'r.-')
plt.title('두번째 그래프')

ax3 = plt.subplot(2, 2, 3)  # ax3 은 2x2 에서 세번째 그래프
plt.plot(x3, y3, 'b--')
plt.title('세번째 그래프')

ax4 = plt.subplot(2, 2, 4)  # ax4 은 2x2 에서 네번째 그래프
plt.plot(x4, y4, 'g.-')
plt.title('네번째 그래프')

plt.show()
```

## 4-3. matplotlib 그래프 유형 별 시각화

### 4-3-1. 바(bar)

```python
# 데이터 생성

x = np.arange(6)
y = [90, 70, 40, 90, 60, 50]
xlabel = ['국어', '수학', '영어', '과학', '미술', '체육']
```

```python
# 세로 막대 그래프
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.ylabel('점수')
plt.show()
```

```python
# 가로 막대 그래프
plt.barh(x, y)
plt.yticks(x, xlabel)
plt.xlabel('점수')
plt.show()
```

막대 그래프 여러개 그리기(나란한 형태)

```python
# 데이터 생성

x = np.arange(5)
women = [80, 70,40, 90, 60]
men = [70, 80, 45, 75, 80]
xlabel = ['국어', '수학', '영어', '과학', '미술']
```

```python
plt.rc('font', size = 10)
bar_width = 0.35   # 막대그래프 폭 설정

plt.bar(x, women, width = bar_width, label = '여자')
plt.bar(x + bar_width, men, width = bar_width, label = '남자')

plt.xticks(x + bar_width/2, xlabel)
plt.ylabel('점수')
plt.legend(loc=0)
plt.show()
```

bar 여러개 그리기(stack 형태)

```python
bar_width = 0.35   # 막대그래프 폭 설정

plt.bar(x, women, width = bar_width, label = '여자', bottom=men)
plt.bar(x, men, width = bar_width, label = '남자')

plt.xticks(x, xlabel)
plt.ylabel('점수')
plt.legend(loc='upper center')
plt.show()
```

### 4-3-2. 파이(pie)

```python
# 데이터 생성

labels = ['애플', '삼성', '오포', '샤오미', '기타']
sizes = [16, 21, 16, 14, 33]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'lightgray']
```

```python
plt.pie(sizes, labels=labels)
plt.title('휴대폰 시장 점유율')
plt.show()
```

```python
# 파이 돌출 옵션

explode = (0, 0.1, 0, 0, 0)

plt.figure(figsize=(5, 5))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%0.1f%%')
plt.title('휴대폰 시장 점유율')
plt.show()
```

3D 그래프(surface)

```python
# 3차원 데이터 생성

from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(50, 150, 24)
y = np.linspace(50, 150, 24)

x, y = np.meshgrid(x, y)
z = (x-100)**2/y
```

```python
fig = plt.figure(figsize = (10, 5))
axes = fig.add_subplot(111, projection='3d')

axes.plot_surface(x, y, z, cmap = 'coolwarm')
plt.title('3D 그래프 시각화')
plt.show()
```
