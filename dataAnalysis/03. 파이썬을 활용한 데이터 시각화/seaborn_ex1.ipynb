### 4-5-1. seaborn의 장점

1. 통계 서포트 기능

```python
# 배경을 darkgrid 로 설정
sns.set(style='darkgrid')

# 데이터 로드
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
```

```python
# 항목의 개수를 세어주는 'countplot'

plt.figure(figsize=(10,6))
sns.countplot(x="class", hue="who", data=titanic)
plt.show()
```

```python
# 선형 회귀 관계를 보여주는 'lmplot'

plt.figure(figsize=(10, 6))
sns.lmplot(x='total_bill', y = 'tip', height = 8, data = tips)
plt.show()
```

```python
# 산점도와 히스토그램 동시에 보여주는 'joinplit'

plt.figure(figsize=(10,6))
sns.jointplot(x="total_bill", y="tip", height=8, data=tips)
plt.show()
```

```python
# 문자열 열 제외
numeric_titanic = titanic.select_dtypes(include=['number'])
numeric_titanic.corr()
```

### 4-5-2. matplotlib 그래프를 seaboen에서 표현하기

```python
# 데이터 로드
penguins = sns.load_dataset("penguins")
penguins.head()
```

matplotlib으로 그래프 그리기

```python
# 2개의 axes를 만들기
fig, axes = plt.subplots(ncols=2, figsize=(10,5))

# axes[0]
islands = penguins['island'].unique()
color = 0

for island in islands:
    x = penguins['bill_length_mm'].loc[penguins['island'] == island]
    y = penguins['bill_depth_mm'].loc[penguins['island'] == island]
    
    axes[0].scatter(x, y, color = f"C{color}", label=island)
    color += 1

axes[0].set_xlabel("bill_length_mm")
axes[0].set_ylabel("bill_depth_mm")

plt.show()
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(10,5))

# ---
# axes[0] : matplotlib
islands = penguins['island'].unique()
color = 0
for island in islands:
    axes[0].scatter(penguins["bill_length_mm"].loc[penguins["island"]==island],
                   penguins["bill_depth_mm"].loc[penguins["island"]==island],
                   c=f"C{color}", label=island)
    color = color + 1

axes[0].set_xlabel("bill_length_mm")
axes[0].set_ylabel("bill_depth_mm")

# ---
# axes[1] : seaborn
sns.scatterplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm",
    hue="island", ax=axes[1]
)


plt.show()
```
